# Feature Selection using SelectKBest

```python
# feature selection function using SelectKBest where k = 5 since we know there are 5 non-redundant features

def feature_selection_skb(X, y):
    skb = SelectKBest(k = 5)
    skb.fit(X, y)
    
    skb_feats = np.where(skb.get_support())[0]
    return skb_feats
```

```python
# creates madelon_train_full from the madelon_train set and also now with the target column

madelon_train_full = madelon_train
madelon_train_full['target'] = madelon_train_targets
```

```python
# starting with a sample size of 100 samples, find the best 5 features for explaining the target column. Continually do this with a step size of 50 until the size of samples is the full data set of 2000 samples
for number_samples in np.arange(100, len(madelon_train_full) + 50, 50):
    ss = StandardScaler()
    
    madelon_train = madelon_train_full.sample(n=number_samples, random_state = 42)
    train_target = madelon_train['target']
    madelon_train = madelon_train.drop('target', 1)
    
    madelon_train_sc = ss.fit_transform(madelon_train)
    
    print("For {} samples of the dataset, the 5 best features are {}".format(number_samples, feature_selection_skb(madelon_train_sc, train_target)))
```

What can be learned from the above code is that the 5 features that are important are most likely 64, 128, 241, 336, and 475. A smaller sample size can return these correct features, however the sample sizes around these smaller sizes do not return the correct features. From my data, it looks like in order for the 5 features to be extracted, a vast majority of the dataset needs to be used in the analysis.

The idea behind the following code is to loop through all of these different sample sizes, determine which 5 SKBest features are chosen using that sample size, and then to run a simple DecisionTreeClassifier and a KNeighborsClassifier using only these 5 features to try to predict the target variable. Using the percent the classifier correctly predicts, it should be relatively easy to see which 5 features are the informative features (of the 20 already determined from above)

```python
# loops through the madelon_train_full dataset with every iteration adding 50 samples to the dataset being tested until eventually the entire dataset is being used.
# uses the SelectKBest function for each size to select the 5 best features from the scaled training data.
final_df = pd.DataFrame()
for number_samples in np.arange(100, len(madelon_train_full) + 50, 50):
    
    ss = StandardScaler()
    
    madelon_train = madelon_train_full.sample(n=number_samples, random_state = 42)
    train_target = madelon_train['target']
    madelon_train = madelon_train.drop('target', 1)
    
    madelon_train_sc = ss.fit_transform(madelon_train)
    
    selected_features = feature_selection_skb(madelon_train_sc, train_target)
    
    # creates the target_selected and train_selected datasets using only the 5 features found from the SKBest function
    train_selected = madelon_train_full[selected_features]
    target_selected = madelon_train_full['target']
    
    # creates the valid_selected and target_valid datasets using only the 5 features found from the SKBest function
    valid_selected = madelon_valid[selected_features]
    target_valid = madelon_val_target
    
    # the code that follows scales the datasets with the selected features and creates 2 DecisionTreeClassifier models and 2 KNeighborsClassifier models, 1 raw data and 1 scaled data.
    train_selected_sc = ss.fit_transform(train_selected)
    valid_selected_sc = ss.transform(valid_selected)
    
    lr = LogisticRegression()
    lr.fit(train_selected, target_selected)
    
    lr_sc = LogisticRegression()
    lr_sc.fit(train_selected_sc, target_selected)
    
    decision_tree = DecisionTreeClassifier(random_state = 42)
    decision_tree.fit(train_selected, target_selected)
    
    decision_tree_sc = DecisionTreeClassifier(random_state = 42)
    decision_tree_sc.fit(train_selected_sc, target_selected)
    
    kneighbors = KNeighborsClassifier()
    kneighbors.fit(train_selected, target_selected)
    
    kneighbors_sc = KNeighborsClassifier()
    kneighbors_sc.fit(train_selected_sc, target_selected)
    
    # the scores for each fit model are calculated using the validation set
    lr_score = lr.score(valid_selected, target_valid)
    decision_tree_score = decision_tree.score(valid_selected, target_valid)
    kneighbors_score = kneighbors.score(valid_selected, target_valid)
    lr_sc_score = lr_sc.score(valid_selected_sc, target_valid)
    decision_tree_sc_score = decision_tree_sc.score(valid_selected_sc, target_valid)
    kneighbors_sc_score = kneighbors_sc.score(valid_selected_sc, target_valid)
    
    scores_list = list()
    
    # add all of the scores to the scores_list
    scores_list.append(lr_score)
    scores_list.append(decision_tree_score)
    scores_list.append(kneighbors_score)
    scores_list.append(lr_sc_score)
    scores_list.append(decision_tree_sc_score)
    scores_list.append(kneighbors_sc_score)
    
    scores_array = np.array(scores_list)
        
    scores_df = pd.DataFrame(scores_array)
    
    final_df[number_samples] = [*scores_array, selected_features]
    
```

```python
final_df = final_df.transpose()
final_df = final_df.rename(columns = {0:'LogisticRegression', 1:'DecisionTreeClassifier', 2:'KNeighborsClassifier', 3:'ScaledLogisticRegression', 4:'ScaledDecisionTreeClassifier', 5:'ScaledKNeighborsClassifier', 6:'SelectedFeatures'})
```
